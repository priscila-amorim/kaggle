```python
sample_sub = pd.read_csv('sample_submission.csv')
df_train_extra = pd.read_csv('Dig-MNIST.csv')
df_train = pd.read_csv('train.csv')
y_train = df_train.iloc[:,0]
X_train = df_train.drop(labels=['label'], axis=1)
df_test = pd.read_csv('test.csv')
X_test = df_test.drop(labels=['id'], axis=1)
```

```python
# preprocessing the images

# convert each image to 1 dimensional array
X = X_train
Y = y_train

# normalize the data to 0 - 1
X = X.astype(float) / 255.
X_test = X_test.astype(float) / 255.
```

```python
t0= time()

n_digits = 400          
print(n_digits)


# Initialize KMeans model
kmeans = MiniBatchKMeans(n_clusters = n_digits, max_iter=50, batch_size = 5120, n_init=77)

kmeans.fit(X)
t1 = time() - t0
print("Time elapsed: ", t1)
```

```python
outcome = kmeans.predict(X_test)
```

```python
def retrieve_info(cluster_labels,y_train):

    # Initializing
    reference_labels = {}
    # For loop to run through each label of cluster label
    for i in range(len(np.unique(kmeans.labels_))):
        index = np.where(cluster_labels == i,1,0)
        num = np.bincount(y_train[index==1]).argmax()
        reference_labels[i] = num
    return reference_labels
```

```python
reference_labels = retrieve_info(kmeans.labels_,y_train)
#number_labels = np.random.rand(len(kmeans.labels_))
value = list()
key = list()
for i in range(len(kmeans.labels_)):
    if len(key)==0:
        key.append(kmeans.labels_[i].tolist())
        value.append(reference_labels[kmeans.labels_[i]].tolist())
    elif kmeans.labels_[i] not in key:
        key.append(kmeans.labels_[i].tolist())
        value.append(reference_labels[kmeans.labels_[i]].tolist())

mapping = dict(zip(key, value))
outcome = pd.DataFrame(outcome)
outcome = outcome.replace(mapping)
```

```python
# os.chdir(r'/kaggle/working/')
sample_sub["label"] =outcome

sample_sub.to_csv('submission.csv', index=False)
```
